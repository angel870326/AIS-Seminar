# Use TSMC and other 9 Companies as an Example

```python
num = [i for i in range(2330,2347)]
companyList = map(str, num)

# companyList = ['2330', '2331', '2332', '2337', '2338', '2340', '2342', '2344', '2345', '2347']
```

```python
import os.path
from os import path
```

```python
company_Code = []
audit_Report = []
# accounting_Firm = []
# CPA_1 = []
# CPA_2 = []

# 6
net_Income = []
total_Equity = []
total_Liabilities = []
total_Assets = []
operating_Cash_Flows = []
basic_EPS = []
# Add 7
total_revenue = []
gross_profit = []
operating_expense = []
profit_loss_from_operating_activities = []
finance_costs = []
income_tax_expense = []
other_comprehensive_income = []

# Add 7



for company in companyList:

    xmlfile = '/content/gdrive/My Drive/碩一下/會計資訊系統研討/tifrs-2018Q4/tifrs-fr1-m1-ci-cr-' + company + '-2018Q4.xml'
    if path.exists(xmlfile):
      tree = ET.parse(xmlfile)
      root = tree.getroot()
    
      auditReport = root.findall('{http://www.xbrl.org/tifrs/ar/2018-09-30}AccountantsReportBody')[0].text
      # accountingFirm = root.findall('{http://www.xbrl.org/tifrs/ar/2018-09-30}AccountantName')[0].text
      # CPA1 = root.findall('{http://www.xbrl.org/tifrs/ar/2018-09-30}AssuranceAccountantName1')[0].text
      # CPA2 = root.findall('{http://www.xbrl.org/tifrs/ar/2018-09-30}AssuranceAccountantName2')[0].text
      netIncome = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}ProfitLoss')[0].text
      totalEquity = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}Equity')[0].text
      totalLiabilities = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}Liabilities')[0].text
      totalAssets = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}Assets')[0].text
      operatingCashFlows = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}CashFlowsFromUsedInOperatingActivities')[0].text
      basicEPS = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}BasicEarningsLossPerShare')[0].text
      # Add
      revenue = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}Revenue')[0].text
      grossProfit = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}GrossProfit')[0].text
      operatingExpense = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}OperatingExpense')[0].text

      profitLossFromOperatingActivities = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}ProfitLossFromOperatingActivities')[0].text
      financeCosts = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}FinanceCosts')[0].text
      incomeTaxExpenseContinuingOperations = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}IncomeTaxExpenseContinuingOperations')[0].text
      otherComprehensiveIncome = root.findall('{http://xbrl.ifrs.org/taxonomy/2017-03-09/ifrs-full}OtherComprehensiveIncome')[0].text
      


      company_Code.append(company)
      # accounting_Firm.append(accountingFirm)
      # CPA_1.append(CPA1)
      # CPA_2.append(CPA2)
      audit_Report.append(auditReport)
      net_Income.append(netIncome)
      total_Equity.append(totalEquity)
      total_Liabilities.append(totalLiabilities)
      total_Assets.append(totalAssets)
      operating_Cash_Flows.append(operatingCashFlows)
      basic_EPS.append(basicEPS)
      # Add
      total_revenue.append(revenue)
      gross_profit.append(grossProfit)
      operating_expense.append(operatingExpense)
      profit_loss_from_operating_activities.append(profitLossFromOperatingActivities)
      finance_costs.append(financeCosts)
      income_tax_expense.append(incomeTaxExpenseContinuingOperations)
      other_comprehensive_income.append(otherComprehensiveIncome)



```

```python
XBRL = pd.DataFrame(company_Code)
XBRL['Company Code'] = pd.DataFrame(company_Code)
XBRL['Revenue'] = pd.DataFrame(total_revenue)
XBRL['Gross Profit'] = pd.DataFrame(gross_profit)
XBRL['Operating Expense'] = pd.DataFrame(operating_expense)
XBRL['Profit from Operating Activities'] = pd.DataFrame(profit_loss_from_operating_activities)
XBRL['Finance Costs'] = pd.DataFrame(finance_costs)
XBRL['Income Tax Expense'] = pd.DataFrame(income_tax_expense)

XBRL['Net Income'] = pd.DataFrame(net_Income)
XBRL['Other Comprehensive Income'] = pd.DataFrame(other_comprehensive_income)

XBRL['Total Equity'] = pd.DataFrame(total_Equity)
XBRL['Total Liabilities'] = pd.DataFrame(total_Liabilities)
XBRL['Total Assets'] = pd.DataFrame(total_Assets)
XBRL['Operating Cash Flows'] = pd.DataFrame(operating_Cash_Flows)
XBRL['Basic EPS'] = pd.DataFrame(basic_EPS)
```
